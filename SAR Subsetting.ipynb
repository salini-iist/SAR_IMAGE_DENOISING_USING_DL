```python
# Creating subsets
k = 0
size_sar = 1024
from PIL import Image

with rasterio.open('s1a-iw-grd-vh-20220430t234945-20220430t235010-043009-05229c-002.tiff') as src:
    for i in range(0,min(im1.profile['width'], im1.profile['height']),size_sar):
        for j in range(0,min(im1.profile['width'], im1.profile['height']),size_sar):
        
            w = src.read(1, window=Window(i,j,size_sar , size_sar))
            outfile = '%s%d%d.tiff'% ("sar_vh_training", i,j)
            w_img = Image.fromarray(w)
            w_img.save(outfile)
            print(f'i={i}, j = {j}')
            k = k + 1

print(f'k = {k}')    
```
